```python
import pandas as pd

marconi = pd.read_parquet("../very_big_data/marconi_full.parquet")
marconi
```

```python
A10_per_hour = 1.2
total_prefill_tokens = marconi['num_input_tokens'].sum()
total_reality_time = 120
total_simulation_time = 0.0305

total_reality_price = total_reality_time * A10_per_hour
total_simulation_price = total_simulation_time * A10_per_hour

print("Total simulation price (EUR):", total_simulation_price)
print("Total reality price (EUR):", total_reality_price)


print("Total prefill tokens:", total_prefill_tokens)
```

```python
def eur_per_Mtok_per_sec(cost_per_hour, hours, tokens):
    cost_eur      = cost_per_hour * hours          # â‚¬
    seconds       = hours * 3600                   # s
    return cost_eur * 1_000_000_000 / tokens / seconds # â‚¬/Mtok/s


A10_per_hour         = 1.2
total_time_reality_h = 120
total_time_simul_h   = 0.0305

tok_prefill = marconi['num_input_tokens'].sum()
tok_decode  = marconi['num_output_tokens'].sum()   # if present
total_tokens = tok_prefill + tok_decode

eur_Mtok_s_real = eur_per_Mtok_per_sec(A10_per_hour,
                                       total_time_reality_h,
                                       total_tokens)
eur_Mtok_s_sim  = eur_per_Mtok_per_sec(A10_per_hour,
                                       total_time_simul_h,
                                       total_tokens)

print(f"Reality    : {eur_Mtok_s_real:>12,.10f} â‚¬/Mtok/s")
print(f"Simulation : {eur_Mtok_s_sim:>12,.10f} â‚¬/Mtok/s")
```
